```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data/_data.csv', index_col=0)
df
```

```python
df['City'] = df['Адрес'].str.split(',').str[0]
# Удалить все строки, где значение в столбце 'City' не равно 'Москва'
df.drop(df[df['City'] != 'Москва'].index, inplace=True)

df = df.drop(['Адрес', "Телефоны", "Описание", "Площадь комнат, м2", "Название ЖК", "Серия дома", "Ссылка на объявление", 'Мусоропровод', 'Высота потолков, м', 'Тип'], axis=1)
```

```python
df['Square'] = df['Площадь, м2'].str.split('/').str[0]
df['Square'] = pd.to_numeric(df['Square'])
df = df.drop(['Площадь, м2'], axis=1)

df['Rooms'] = df['Количество комнат'].str.split(',').str[0]
df['Rooms'] = pd.to_numeric(df['Rooms'])
df = df.drop(['Количество комнат'], axis=1)
```

```python
def remove_outliers(group, column):
    Q1 = group[column].quantile(0.25)
    Q3 = group[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return group[(group[column] >= lower_bound) & (group[column] <= upper_bound)]

df_square_cleaned = df.groupby('Rooms').apply(remove_outliers, 'Square').reset_index(drop=True)

square_mean = df_square_cleaned.groupby('Rooms')['Square'].mean()
square_mean
```

```python
def fill_missing_rooms(row):
    if pd.isna(row['Rooms']):
        
        closest_room_count = square_mean.sub(row['Square']).abs().idxmin()
        return closest_room_count
    else:
        return row['Rooms']

df['Rooms'] = df.apply(fill_missing_rooms, axis=1)
```

```python
df['Clear_price'] = df['Цена'].str.split().str[0]
df['Clear_price'] = pd.to_numeric(df['Clear_price'])
df = df.drop(['Цена'], axis=1)

df_price_clean = df.groupby('Ремонт').apply(remove_outliers, 'Clear_price').reset_index(drop=True)

price_mean = df_price_clean.groupby('Ремонт')['Clear_price'].mean()

def fill_missing_renovation(row):
    if pd.isna(row['Ремонт']):
        
        closest_renov_count = price_mean.sub(row['Clear_price']).abs().idxmin()
        return closest_renov_count
    else:
        return row['Ремонт']


df['Ремонт'] = df.apply(fill_missing_renovation, axis=1)
```

```python
df.info()
```

```python
df['Окна'] = df['Окна'].fillna('Во двор')
df.nunique()
```

```python
df['Балкон'] = df['Балкон'].fillna('Балкон (0), Лоджия (0)')
df['Лоджия'] = [row.split(' ')[-1] if (row.split(' ')[-2]) == 'Лоджия' else '(0)' for row in df['Балкон']]
df['Балкон'] = [row.split(' ')[1] if row.split(' ')[0] == 'Балкон' else '(0)' for row in df['Балкон']]
```

```python
df['Лоджия'] = [row.split('(')[1] for row in df['Лоджия']]
df['Лоджия'] = [int(row.split(')')[0]) for row in df['Лоджия']]
df['Балкон'] = [row.split('(')[1] for row in df['Балкон']]
df['Балкон'] = [int(row.split(')')[0]) for row in df['Балкон']]
```

```python

df['Этажность дома'] = [int(row.split('/')[1].split(',')[0]) for row in df['Дом']]
df['Этаж квартиры'] = [int(row.split('/')[0]) for row in df['Дом']]
df = df.drop(['Дом'], axis = 1)
```

```python
elevator = df.groupby('Лифт')['Лифт'].count()

sns.barplot(y=elevator.index, x=elevator.values)
plt.ylabel('Тип и количество лифтов')
plt.xlabel('Количество предложений')
plt.title('Большое количество домов только с пассажирским лифтом');
```

```python
df['Лифт'] = [row if row == row else
              ['Пасс (1), Груз (1)' if df['Этажность дома'].iloc[i]>5 else 'Пасс (0), Груз (0)'][0]
              for i, row in enumerate(df['Лифт'])]
              

df['Лифт пасс'] = [row.split(' ')[1] if (row.split(' ')[0]) == 'Пасс' else '(0)' for row in df['Лифт']]
df['Лифт груз'] = [row.split(' ')[-1] if (row.split(' ')[-2]) == 'Груз' else '(0)' for row in df['Лифт']]

df['Лифт пасс'] = [row.split('(')[1] for row in df['Лифт пасс']]
df['Лифт пасс'] = [int(row.split(')')[0]) for row in df['Лифт пасс']]
df['Лифт груз'] = [row.split('(')[1] for row in df['Лифт груз']]
df['Лифт груз'] = [int(row.split(')')[0]) for row in df['Лифт груз']]
df = df.drop(['Лифт'], axis = 1)
```

```python
bath_df = df.groupby('Санузел').size().sort_values(ascending=False).reset_index(name='Count')
bath_df.plot(kind='barh', x='Санузел', legend=None, xlabel='Количество объявлений', ylabel='Тип', title='С большим отрывом доминируют варианты с совмещенным санузлом');
```

```python
mask = df['Санузел'].isna()
df.loc[mask, 'Санузел'] = np.random.choice(['Совмещенный (1)', 'Раздельный (1)'], size=mask.sum())

df['Санузел совмещенный'] = [row.split(' ')[1] if (row.split(' ')[0]) == 'Совмещенный' else '(0)' for row in df['Санузел']]
df['Санузел раздельный'] = [row.split(' ')[-1] if (row.split(' ')[-2]) == 'Раздельный' else '(0)' for row in df['Санузел']]
df['Санузел совмещенный'] = [row.split('(')[1] for row in df['Санузел совмещенный']]
df['Санузел совмещенный'] = [int(row.split(')')[0]) for row in df['Санузел совмещенный']]
df['Санузел раздельный'] = [row.split('(')[1] for row in df['Санузел раздельный']]
df['Санузел раздельный'] = [int(row.split(')')[0]) for row in df['Санузел раздельный']]
df = df.drop(['Санузел'], axis = 1)
```

```python
df['Можно с детьми/животными'] = df['Можно с детьми/животными'].fillna('Нельзя')
```
