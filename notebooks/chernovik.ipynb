```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('_data.csv', index_col=0)
```

```python
df['Метро'] = [row.split(' (')[0] for row in df['Метро'].astype(str)]
df['Метро'] = df['Метро'].str.split('. ')
df
```

```python

df['Square'] = df['Площадь, м2'].str.split('/').str[0]

df['Square'] = pd.to_numeric(df['Square'])

df['Rooms'] = df['Количество комнат'].str.split(',').str[0]

df['Rooms'] = pd.to_numeric(df['Rooms'])
```

```python
# Избавляемся от выбросов

def remove_outliers(group, column):
    Q1 = group[column].quantile(0.25)
    Q3 = group[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return group[(group[column] >= lower_bound) & (group[column] <= upper_bound)]

df_cleaned = df.groupby('Rooms').apply(remove_outliers, 'Square').reset_index(drop=True)


mean = df_cleaned.groupby('Rooms')['Square'].mean()
mean
```

```python
def fill_missing_rooms(row):
    if pd.isna(row['Rooms']):
        
        closest_room_count = mean.sub(row['Square']).abs().idxmin()
        return closest_room_count
    else:
        return row['Rooms']


df['Rooms'] = df.apply(fill_missing_rooms, axis=1)

```

```python
df['Clear_price'] = df['Цена'].str.split().str[0]
df['Clear_price'] = pd.to_numeric(df['Clear_price'])
df

df_price_clean = df.groupby('Ремонт').apply(remove_outliers, 'Clear_price').reset_index(drop=True)
df_price_clean

price_mean = df_price_clean.groupby('Ремонт')['Clear_price'].mean()
price_mean

def fill_missing_renovation(row):
    if pd.isna(row['Ремонт']):
        
        closest_renov_count = price_mean.sub(row['Clear_price']).abs().idxmin()
        return closest_renov_count
    else:
        return row['Ремонт']


df['Ремонт'] = df.apply(fill_missing_renovation, axis=1)
```
