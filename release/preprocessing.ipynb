```python
def plot_mean_price(column):
    plt.figure(figsize=(14, 7)) 
    n_bars = np.arange(len(df[column].unique()))
    grouped = pd.DataFrame(df.groupby([column])['Clear_price'].mean().round())
    
    plt.title(column) 
    plt.ylabel('Цена квартиры')
    plt.bar(n_bars, grouped['Clear_price']) 
    plt.xticks(n_bars, grouped.index); 


categorical = ['parking', 'renovation', 'kids_pets_allowed', 'Rooms', 'house_floor', 'appt_floor', 'balcony_and_loggy', 'balcony_and_loggy', 'sum_bath']

for col in categorical:
    plot_mean_price(col)
```

```python
mean_values = df.groupby('Rooms')['Clear_price'].mean()
median_values = df.groupby('Rooms')['Clear_price'].median()

summary_df = pd.DataFrame({
    'Средняя цена': mean_values,
    'Медианная цена': median_values
}).reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='Rooms', y='value', hue='Метрики', 
            data=pd.melt(summary_df, id_vars='Rooms', value_vars=['Средняя цена', 'Медианная цена'], var_name='Метрики'))
plt.title('Средняя и медианная цена по каждому типу квартиры')
plt.xlabel('Количество комнат')
plt.ylabel('Цена тыс.руб')

plt.grid(True)

median_values
```
