```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('_data.csv', index_col=0)
df.head(5)
```

```python
a = df.groupby('Тип').size()
b = df.groupby('Количество комнат').size()
c = df.groupby('Название ЖК').size().sort_values(ascending=False).reset_index()

```

```python
df['Первая_площадь'] = df['Площадь, м2'].str.split('/').str[0]

df['Первая_площадь'] = pd.to_numeric(df['Первая_площадь'])

df['Количество_комнат_без_типов'] = df['Количество комнат'].str.split(',').str[0]

df['Количество_комнат_без_типов'] = pd.to_numeric(df['Количество_комнат_без_типов'])



df
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Первая_площадь', y='Количество_комнат_без_типов', alpha=0.6)
plt.title('Зависимость площади квартир от количества комнат')
plt.xlabel('Площадь квартиры (кв.м)')
plt.ylabel('Количество комнат')
plt.grid(True)
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Первая_площадь'])
plt.title('Boxplot площади квартир')
plt.xlabel('Площадь квартиры (кв.м)')
plt.grid(True)
```
